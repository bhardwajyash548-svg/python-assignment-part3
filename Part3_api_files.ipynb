```python
# Task 1 — File Read & Write Basics

# -----------------------------
# Part A — Write 
# -----------------------------

try:
    # Writing initial content to file
    with open("python_notes.txt", "w", encoding="utf-8") as file:
        file.write("Topic 1: Variables store data. Python is dynamically typed.\n")
        file.write("Topic 2: Lists are ordered and mutable.\n")
        file.write("Topic 3: Dictionaries store key-value pairs.\n")
        file.write("Topic 4: Loops automate repetitive tasks.\n")
        file.write("Topic 5: Exception handling prevents crashes.\n")
    
    print("File written successfully.")

    # Appending extra lines
    with open("python_notes.txt", "a", encoding="utf-8") as file:
        file.write("Topic 6: Functions help in code reusability.\n")
        file.write("Topic 7: File handling allows data storage.\n")

    print("Lines appended successfully.")

except Exception as e:
    print("Error while writing to file:", e)


# -----------------------------
# Part B — Read from File
# -----------------------------

try:
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()

    print("\n--- File Content ---")

    # Print numbered lines
    for i, line in enumerate(lines, start=1):
        print(f"{i}. {line.strip()}")

    # Count total lines
    print("\nTotal number of lines:", len(lines))

    # Search functionality
    keyword = input("\nEnter a keyword to search: ").lower()
    
    found = False
    print("\nSearch Results:")

    for line in lines:
        if keyword in line.lower():
            print(line.strip())
            found = True

    if not found:
        print("No matching lines found.")

except FileNotFoundError:
    print("File not found. Please make sure the file exists.")
except Exception as e:
    print("Error while reading file:", e)
```

```python
# Task 2 — API Integration

import requests

# -----------------------------
# Step 1 — Fetch and Display Products
# -----------------------------

try:
    url = "https://dummyjson.com/products?limit=20"
    response = requests.get(url)

    data = response.json()
    products = data["products"]

    print("\n--- Product List ---")
    print("ID | Title                     | Category      | Price   | Rating")
    print("-" * 70)

    for p in products:
        print(f"{p['id']}  | {p['title'][:25]:25} | {p['category'][:12]:12} | ${p['price']} | {p['rating']}")

except Exception as e:
    print("Error fetching products:", e)


# -----------------------------
# Step 2 — Filter and Sort
# -----------------------------

try:
    # Filter products with rating >= 4.5
    filtered = [p for p in products if p["rating"] >= 4.5]

    # Sort by price (descending)
    sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)

    print("\n--- Filtered & Sorted Products (Rating >= 4.5) ---")

    for p in sorted_products:
        print(f"{p['title']} - Price: ${p['price']} - Rating: {p['rating']}")

except Exception as e:
    print("Error in filtering/sorting:", e)


# -----------------------------
# Step 3 — Search by Category
# -----------------------------

try:
    laptop_url = "https://dummyjson.com/products/category/laptops"
    response = requests.get(laptop_url)

    laptops = response.json()["products"]

    print("\n--- Laptops Category ---")

    for item in laptops:
        print(f"{item['title']} - Price: ${item['price']}")

except Exception as e:
    print("Error fetching laptops:", e)


# -----------------------------
# Step 4 — POST Request (Simulated)
# -----------------------------

try:
    post_url = "https://dummyjson.com/products/add"

    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = requests.post(post_url, json=new_product)

    print("\n--- POST Response ---")
    print(response.json())

except Exception as e:
    print("Error in POST request:", e)
```

```python
# Task 3 — Exception Handling

import requests

# -----------------------------
# Part A — Guarded Calculator
# -----------------------------

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# Testing
print("\n--- Safe Divide Tests ---")
print(safe_divide(10, 2))       # valid
print(safe_divide(10, 0))       # zero division
print(safe_divide("ten", 2))    # wrong type


# -----------------------------
# Part B — Guarded File Reader
# -----------------------------

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

# Testing
print("\n--- File Reader Test ---")
print(read_file_safe("python_notes.txt"))   # should work
print(read_file_safe("ghost_file.txt"))     # should fail


# -----------------------------
# Part C — Robust API Calls
# -----------------------------

def fetch_products():
    try:
        url = "https://dummyjson.com/products?limit=20"
        response = requests.get(url, timeout=5)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


def create_product():
    try:
        url = "https://dummyjson.com/products/add"
        data = {
            "title": "My Custom Product",
            "price": 999,
            "category": "electronics",
            "description": "A product I created via API"
        }
        response = requests.post(url, json=data, timeout=5)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)

# Test API functions
print("\n--- API Test ---")
print(fetch_products())
print(create_product())


# -----------------------------
# Part D — Input Validation Loop
# -----------------------------

print("\n--- Product Lookup ---")

while True:
    user_input = input("Enter a product ID (1–100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate input
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue

    # API Call
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            product = response.json()
            print(f"Title: {product['title']}, Price: ${product['price']}")
        else:
            print("Unexpected response:", response.status_code)

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
# Task 4 — Logging to File

import requests
from datetime import datetime

# -----------------------------
# Logger Function
# -----------------------------

def log_error(function_name, error_type, message):
    try:
        with open("error_log.txt", "a", encoding="utf-8") as log:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            log.write(f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n")
    except Exception as e:
        print("Logging failed:", e)


# -----------------------------
# 1. Trigger ConnectionError
# -----------------------------

try:
    bad_url = "https://this-host-does-not-exist-xyz.com/api"
    response = requests.get(bad_url, timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Connection error triggered.")
    log_error("fetch_products", "ConnectionError", str(e))


# -----------------------------
# 2. Trigger HTTP Error (404)
# -----------------------------

try:
    url = "https://dummyjson.com/products/999"  # non-existing ID
    response = requests.get(url, timeout=5)

    if response.status_code != 200:
        print("HTTP error triggered.")
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID 999"
        )

except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))


# -----------------------------
# 3. Display Log File Content
# -----------------------------

try:
    print("\n--- Error Log Content ---")
    with open("error_log.txt", "r", encoding="utf-8") as log:
        print(log.read())
except FileNotFoundError:
    print("Log file not found.")
```
